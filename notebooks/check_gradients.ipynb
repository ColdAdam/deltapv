```python
import jaxpv
import jax.numpy as np
from jax import grad, jacfwd, jit, ops
```

```python
N = 500
material = jaxpv.Material()["Si"]
material["A"] = 1e4
grid = np.linspace(0, 3e-4, N)
sim = jaxpv.JAXPV(grid)
sim.add_material(material, np.arange(N))
sim.contacts(1e7, 0, 0, 1e7)
sim.single_pn_junction(1e17, -1e15, 50e-7)
# phi = 1e17
# alpha = 2.3e4
# G = phi * alpha * np.exp(-alpha * grid)
# sim.optical_G("user", G)
sim.incident_light(kind="sun")
sim.optical_G(kind="direct")
gradeff = grad(jaxpv.efficiency)
```

```python
eff = jaxpv.efficiency(sim.data, sim.opt)
gradients = gradeff(sim.data, sim.opt)
```

```python
epsilon = .1
sim.data["Eg"] = ops.index_update(sim.data["Eg"], ops.index[0], sim.data["Eg"][1] + epsilon)
eff_new = jaxpv.efficiency(sim.data, sim.opt)
```

```python
print(gradients["Eg"][0])
print((eff_new - eff) / epsilon)
```
